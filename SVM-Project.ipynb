```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
train_data = pd.read_csv('diabetes.csv')
test_data = pd.read_csv('diabetes_test_unlabeled.csv')
sample_data = pd.read_csv('Sample_Submission.csv')
```

```python
train_data.shape
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
sample_data.head()
```

```python
train_data.drop('encounter_id',axis =1,inplace = True)
train_data.drop('hospital_id',axis =1,inplace = True)
train_data.drop('icu_id',axis =1,inplace = True)
train_data.drop('nan_counts',axis =1,inplace = True)
train_data.drop('d1_row_nan_counts',axis =1,inplace = True)
train_data.drop('h1_row_nan_counts',axis =1,inplace = True)
```

```python
for i in train_data.columns:
    missing = train_data[i].isnull().mean() * 100
    print(f"{i} - {missing:.2f}%")
```

```python
train_data = train_data.dropna(thresh=len(train_data) * 0.55, axis=1)
```

```python
for i in train_data.columns:
    missing = train_data[i].isnull().mean() * 100
    print(f"{i} - {missing:.2f}%")
```

```python
test_data.drop('encounter_id',axis =1,inplace = True)
test_data.drop('hospital_id',axis =1,inplace = True)
test_data.drop('icu_id',axis =1,inplace = True)
test_data.drop('nan_counts',axis =1,inplace = True)
test_data.drop('d1_row_nan_counts',axis =1,inplace = True)
test_data.drop('h1_row_nan_counts',axis =1,inplace = True)
```

```python
for i in test_data.columns:
    missing = test_data[i].isnull().mean() * 100
    print(f"{i} - {missing:.2f}%")
```

```python
test_data = test_data.dropna(thresh=len(test_data) * 0.55, axis=1)
```

```python
for i in test_data.columns:
    missing = test_data[i].isnull().mean() * 100
    print(f"{i} - {missing:.2f}%")
```

```python
train_data.shape[1]
test_data.shape[1]
```

```python
train_data.head()
```

```python
train_data.select_dtypes(include='object').isnull().any()
train_data['ethnicity'] = train_data['ethnicity'].fillna(train_data['ethnicity'].mode()[0])
train_data['gender'] = train_data['gender'].fillna(train_data['gender'].mode()[0])
train_data['hospital_admit_source'] = train_data['hospital_admit_source'].fillna(train_data['hospital_admit_source'].mode()[0])
train_data['icu_admit_source'] = train_data['icu_admit_source'].fillna(train_data['icu_admit_source'].mode()[0])
```

```python
train_data.select_dtypes(include='object').isnull().any()
```

```python
test_data.select_dtypes(include='object').isnull().any()
test_data['ethnicity'] = test_data['ethnicity'].fillna(test_data['ethnicity'].mode()[0])
test_data['gender'] = test_data['gender'].fillna(test_data['gender'].mode()[0])
test_data['hospital_admit_source'] = test_data['hospital_admit_source'].fillna(test_data['hospital_admit_source'].mode()[0])
test_data['icu_admit_source'] = test_data['icu_admit_source'].fillna(test_data['icu_admit_source'].mode()[0])
```

```python
imputer = SimpleImputer(strategy='mean')
imputer.fit(train_data.select_dtypes(include=['number']))
filled = pd.DataFrame(imputer.transform(train_data.select_dtypes(include=['number'])), columns=train_data.select_dtypes(include=['number']).columns)
train_data = pd.concat([filled, train_data.select_dtypes(exclude=['number'])], axis=1)

imputer = SimpleImputer(strategy='mean')
imputer.fit(test_data.select_dtypes(include=['number']))
filled2 = pd.DataFrame(imputer.transform(test_data.select_dtypes(include=['number'])), columns=test_data.select_dtypes(include=['number']).columns)
test_data = pd.concat([filled2, test_data.select_dtypes(exclude=['number'])], axis=1)
```

```python
train_missing_columns = train_data.columns[train_data.isnull().any()]
test_missing_columns = test_data.columns[test_data.isnull().any()]

print(train_missing_columns)
print(test_missing_columns)
```

```python
X_train = train_data.drop('diabetes_mellitus', axis=1)
y_train = train_data['diabetes_mellitus']
```

```python
X_train.shape
```

```python
scaler = StandardScaler()
scaler.fit(X_train.select_dtypes(include=['number']))
X_train_standardized = pd.DataFrame(scaler.transform(X_train.select_dtypes(include=['number'])), columns=X_train.select_dtypes(include=['number']).columns)
X_train = pd.concat([X_train_standardized, X_train.select_dtypes(exclude=['number'])], axis=1)

scaler.fit(test_data.select_dtypes(include=['number']))
test_data_standardized = pd.DataFrame(scaler.transform(test_data.select_dtypes(include=['number'])), columns=test_data.select_dtypes(include=['number']).columns)
test_data = pd.concat([test_data_standardized, test_data.select_dtypes(exclude=['number'])], axis=1)
```

```python
X_train = pd.DataFrame(X_train)
X_train.head()
```

```python
X_train = pd.get_dummies(X_train)
test_data = pd.get_dummies(test_data)
```

```python
X_train.head()
test_data.head()
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.3, random_state=0)
```

```python
X_train
y_train
```

```python
svm = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=0)
svm.fit(X_train, y_train)
```

```python
y_pred = svm.predict(X_val)
```

```python
print(y_pred.shape)
print(y_train)
```

```python
accuracy = accuracy_score(y_val, y_pred)
accuracy
```

```python
pred = svm.predict(test_data)
```

```python
pred_df = pd.DataFrame(pred,columns=['diabetes_mellitus'])
```
